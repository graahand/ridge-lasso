```python
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=load_boston()
```

```python
df
```

```python
# converting aray into dataset
dataset = pd.DataFrame(df.data)
```

```python
dataset.head()
```

```python
dataset.columns=df.feature_names
dataset.head()
# assigning features name
```

```python
#creating a dependent variable
df.target.shape
```

```python
dataset['Price']=df.target
dataset
```

```python
X=dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
# X is independent feature
# y is dependent feature
```

# Ridge and Lasso regression

```python
# ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
model = Ridge()
parameters = {'alpha':[1e-15,1e-10,1,5,10,100,2000,10000]}
ridge_regressor_model= GridSearchCV(model,parameters,cv=5)
ridge_regressor_model.fit(X,y)
print(ridge_regressor_model.best_params_)
print(ridge_regressor_model.best_score_)
```

```python
# lasso regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso_model = Lasso()
parameters={'alpha':[1,100,10,12,17,5,6,9,89,80]}
lasso_regressor=GridSearchCV(lasso_model,parameters,cv=5)
lasso_regressor.fit(X,y)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```
